```python
import pandas as pd

summary_rows = []

for name, df in datasets.items():
    summary_rows.append({
        "dataset": name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "total_missing_cells": int(df.isna().sum().sum())
    })

summary_df = pd.DataFrame(summary_rows).sort_values("rows", ascending=False)
summary_df
```
